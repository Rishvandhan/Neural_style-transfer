```python
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = "D:/python/workspace_machineLearning/Style transfer/saved_model.pb""
```

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3 )
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
content_image = load_image('DSC_0140.JPG')
style_image = load_image('_17f150fb-7cec-4a33-8b01-afa5b1d5cb3a.jpg')
```
